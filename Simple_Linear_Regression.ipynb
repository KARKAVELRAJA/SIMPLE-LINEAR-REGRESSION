## ***IMPORTING LIBRARIES***

```python
import numpy as np
import pandas as pd
```

## ***LOADING THE DATASET***

```python
data = pd.read_csv("Housing.csv")
```

## ***SUMMARIZING THE DATASET***

```python
print(data.shape)
print(data.head())
```

## ***SEGREGATING THE DATASET INTO INPUT AND OUTPUT***

```python
input  = data.area.values
print(input)
output = data.price.values  
print(output)
```

## ***VISUALIZING THE DATASET***

```python
import matplotlib.pyplot as plt
plt.xlabel("AREA")
plt.ylabel("PRICE")
plt.scatter(input,output,marker="*")
```

## ***TRAINING THE DATASET***

```python
from sklearn.linear_model import LinearRegression

Simple_Linear_Regression_Model = LinearRegression()

Simple_Linear_Regression_Model.fit(input.reshape(-1,1),output.reshape(-1,1))
```

## ***PREDICTING THE OUTPUT USING THE TRAINED DATASET***

```python
X = 7420

print("THE PRICE OF THE HOUSE WITH AREA",X,"SQUARE FEET IS",Simple_Linear_Regression_Model.predict([[X]]))
```

## ***CHECKING THE PREDICTED OUTPUT USING Y=MX+C***

```python
M = Simple_Linear_Regression_Model.coef_
C = Simple_Linear_Regression_Model.intercept_

Y = (M*X)+C

print("THE PRICE OF THE HOUSE WITH AREA",X,"SQUARE FEET IS",Y)
```
